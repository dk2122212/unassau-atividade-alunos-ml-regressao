```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from IPython.display import display

# 2. Caminho do arquivo students_performance.csv
caminho_arquivo = '../data/datasets/students_performance.csv' 
df = pd.read_csv(caminho_arquivo)

# 3. Análise Rápida de Visão Geral
print("--- Primeiras 5 Linhas (PARA IDENTIFICAR COLUNAS) ---")
display(df.head()) 

print("\n--- Formato do DataFrame (Linhas, Colunas) ---")
print(df.shape)

print("\n--- Tipos de Dados e Valores Não-Nulos ---")
df.info()

print("\n--- Estatísticas Descritivas ---")
display(df.describe(include='all').T) 

print("\n--- Contagem de Valores Faltantes por Coluna ---")
print(df.isnull().sum())
```

```python
target_column = 'final_grade' 
print(f"Analisando a Distribuição da Variável Target: {target_column}")

plt.figure(figsize=(15, 5))

# 1. Distribuição (Histograma)
plt.subplot(1, 2, 1)
sns.histplot(df[target_column], kde=True)
plt.title(f'Distribuição da {target_column}')
plt.xlabel(target_column)

# 2. Boxplot (para identificar Outliers)
plt.subplot(1, 2, 2)
sns.boxplot(y=df[target_column])
plt.title(f'Boxplot para {target_column} (Outliers)')

plt.tight_layout()
plt.show()
```

```python
target_column = 'final_grade' # Já identificada na análise anterior

# 1. Calcula a matriz de correlação apenas para colunas numéricas
# O parâmetro numeric_only=True é importante, pois 'final_grade' é numérica.
correlation_matrix = df.corr(numeric_only=True)

print("--- 5 Variáveis mais correlacionadas com a Final_Grade ---")
# Filtra a coluna 'final_grade' da matriz e ordena do maior para o menor.
# Ignoramos o primeiro valor, que é 1.00 (a correlação de final_grade com ela mesma).
print(correlation_matrix[target_column].sort_values(ascending=False).head(5))

# 2. Visualização da Matriz de Correlação (Heatmap)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=.5)
plt.title('Mapa de Calor da Correlação')
plt.show()
```
